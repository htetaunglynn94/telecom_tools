```python
import pandas as pd
import numpy as np
import os
files = os.listdir()
sheets = ['GSM Carrier', 'RU(GSM)']
for s in sheets:
    df = pd.DataFrame()
    for f in files:
        if f[-4::] == 'xlsm':
            d = pd.read_excel(f, sheet_name=s)
            print(f+' file is read.')
            d = d.drop(index=[0,1,2,3]).reset_index(drop=True)
            df = df.append(d, ignore_index=True)
    df.to_excel('ZTE RNC File Combiner_'+s+'.xlsx', index=False)
print('Operation completed')
```
