```python
import folium 
import pandas as pd
from folium import plugins
```

```python
data = 'CCTracker.xlsx'
cc_df = pd.read_excel(data)
cc_df.columns
```

```python
latitudes = list(cc_df.Latitude)
longitudes = list(cc_df.Longitude)
CC_num = list(cc_df.Complaint_Number)
```

```python
print(cc_df.Latitude.dtype)
print(cc_df.Longitude.dtype)
print(cc_df.Complaint_Number.dtype)
```

```python
cc_map = folium.Map(location=[16.842358, 96.159029],zoom_start=12)
```

```python
Mkr_Cls = plugins.MarkerCluster().add_to(cc_map)
```

```python
for lat, lng, label in zip(latitudes, longitudes, CC_num):
    folium.Marker(
                location = [lat,lng],
                icon=None,
                popup=label).add_to(Mkr_Cls)

cc_map
```

```python
cc_map.save('CCTracker.html')
```
