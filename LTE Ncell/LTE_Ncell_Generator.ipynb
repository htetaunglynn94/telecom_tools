```python
import pandas as pd

ncell = 'LTE_Ncell.csv'
vlcell = 'VL.xlsx'

dfvl = pd.read_excel(vlcell)
print('Cell list imported.')

#conn = sq3.connect(ncell)
#cursor = conn.cursor()
#dfdb = pd.read_sql_query('SELECT * FROM LTE_Ncell', conn)

dfdb = pd.read_csv(ncell, skiprows=6, skipfooter=1, engine='python')
print('LTE Ncell data imported.')
result = dfdb.merge(dfvl, how='right', on = 'Local cell name')

filter = result.loc[:,['eNodeB Name', 'Local cell name', 'Local cell identity', 'Target Cell Name', 'Target Cell ID', 'Target eNodeB ID', 'L.HHO.NCell.ExecAttOut']]
filter = filter.fillna(0)
filter['L.HHO.NCell.ExecAttOut'] = filter['L.HHO.NCell.ExecAttOut'].astype(int)

result1 = filter.sort_values(by=['Local cell name', 'L.HHO.NCell.ExecAttOut'], 
           ascending = [True,False], ignore_index=True)

count = len(result1.index)

#result1 = result1.reset_index()
result1.loc[0,'Rank'] = 1

for i in range(1,count):
    if result1.loc[i,'Local cell name'] == result1.loc[i-1,'Local cell name']:
      result1.loc[i,'Rank'] = result1.loc[i-1,'Rank'] + 1
    else: result1.loc[i,'Rank'] = 1

result1['Rank'] = result1['Rank'].astype(int)
result1.to_excel('LTE_Ncell_Final.xlsx', index=False)
print('Operation complated')
```
