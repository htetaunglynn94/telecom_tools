```python
df_cmop = df_cm[(df_cm['Vendor']=='HUAWEI') & (df_cm['Update Category']=='For OPT')].reset_index(drop=True)
df_cmop
```

```python
len(df_cmop['PIC'])
```

```python
pic_name = {0:'Htet Aung Lynn', 1:'Thida San', 2:'Aye Thandar', 3:'Lian Suan Khai'}
```

```python
random_num = list(np.linspace(start=0, stop=4, num=df_cmop.shape[0], endpoint=False, dtype='int'))
```

```python
len(random_num)
```

```python
np.random.shuffle(random_num)
```

```python
random_num
```

```python
df_cmop['PIC'] = random_num
pics = list(df_cmop['PIC'].unique())
pics
```

```python
df_cmop['PIC'].replace(pic_name, inplace=True)
```

```python
df_cmop['PIC'].value_counts()
```

```python
df_cmop
```

```python
cell_counts = pd.DataFrame(df_cmop.groupby('PIC')['PIC'].count()).rename(columns={'PIC':'Cell Counts'}).reset_index()
```

```python
with pd.ExcelWriter('SPM_Low_Thpt_Distribution by HAL.xlsx') as writer:
    cell_counts.to_excel(writer, sheet_name='Summary', index=False)
    df_cmop.to_excel(writer, sheet_name='Assignment', index=False)
```
